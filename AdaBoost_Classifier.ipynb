```python
import pandas as pd
from sklearn.datasets import make_classification
from  sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
x , y = make_classification(n_samples = 1000 , n_features = 5 , n_classes = 2 , random_state = 42)
```

```python
x
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.3 , random_state = 42)
```

```python
basemod = DecisionTreeClassifier(max_depth = 1)
```

```python
adamod = AdaBoostClassifier(estimator = basemod , n_estimators = 200 , random_state = 42)
```

```python
adamod.fit(x_train , y_train)
```

```python
y_pred = adamod.predict(x_test)
```

```python
print('Training score: ', adamod.score(x_train,y_train))
```

```python
print('Testing score: ', adamod.score(x_test,y_test))
```

```python
print('Precision: ', precision_score(y_test,y_pred))
```

```python
confusion_mat=confusion_matrix(y_test,y_pred)
confusion_mat
```

```python
print('Recall : ', recall_score(y_test,y_pred))
```

```python
print('F1 : ', f1_score(y_test,y_pred))
```
