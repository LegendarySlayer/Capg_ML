#  Phone Number

```python
df['Phone_Number'].value_counts().head(20)
```

```python
df['Cleaned_Phone'] = df['Phone_Number'].astype(str).replace(r'\D' , '' ,regex = True)
```

```python
# Extract country code (first 1 - 3 dgit if number >  10 digits)
def get_country_code(number):
    if pd.isnull(number):
        return np.nan
    if len(number) <= 10:
        return np.nan
    return number[:-10] # everything before last 10 digit
df['Country_Code'] = df['Cleaned_Phone'].apply(get_country_code)
```

```python
df['Country_Code'].isnull().sum()
```

```python
def get_local_number(number):
    if pd.isnull(number) or len(number) < 10:
        return np.nan
    return number[-10:]
df['Local_Number'] = df['Cleaned_Phone'].apply(get_local_number)
```

```python
df['Is_Valid_Number'] = df['Local_Number'].notnull()
```

```python
def salary_forest_type(s):
    s=str(s).strip()
    if s=='nan':
        return 'Missing'
    elif any(c.isalpha() for c in s):
        return 'Contains letters (k, M , etc.)'
    elif any(c in '$5' for c in s):
        return 'Currency / coma'
    elif s.replace('.','',1).isdigit():
        return 'Numeric'
    else:
        return 'Other / invalid'
Salary_Forest = salary_series.apply(salary_forest_type)

#check counts per forest type
Salary_Forest.value_counts()
```
