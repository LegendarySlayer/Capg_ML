```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn import datasets
dir(datasets)
```

```python
# Creating dataset
data = {
    "Name": ["Aman","Riya","Rahul","Sneha","Arjun","Pooja","Karan","Neha","Vikram","Anjali"],
    "Gender": ["Male","Female","Male","Female","Male","Female","Male","Female","Male","Female"],
    "Department": ["CSE","IT","ECE","CSE","ME","IT","CSE","ECE","ME","CSE"],
    "City": ["Delhi","Mumbai","Kolkata","Delhi","Chennai","Mumbai","Delhi","Kolkata","Chennai","Delhi"],
    "Performance": ["Good","Average","Poor","Good","Average","Good","Good","Average","Poor","Good"]
}

df = pd.DataFrame(data)

print("Dataset Created Successfully:\n")
print(df)
```

label Emcoding

```python
le = LabelEncoder()
df["Gender"] = le.fit_transform(df["Gender"])
# df["Department"] = le.fit_transform(df["Department"])
df
```

OneHot Encoding

```python
oh = OneHotEncoder(sparse_output = False)
data = oh.fit_transform(df[["Department"]])
name = oh.get_feature_names_out()
encode_df = pd.DataFrame(data,columns = name)
encode_df
```

```python
# concat the new data frame
df = pd.concat([df, encode_df], axis=1)
df
```

```python
# dropping the dept colmn

df.drop("Department", axis=1, inplace=True)
df
#
```

Ordinal Encoding

```python
oe = OrdinalEncoder(categories=[["Poor","Average","Good"]])
df["Performance_encoded"] = oe.fit_transform(df[["Performance"]])
df
```
