```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/Jeevita-R/ML_datasets/raw/refs/heads/main/House_price_prediction.csv")
```

```python
df
```

```python
df['Suburb'].unique()
df['Type'].unique() #label
df['Method'].unique() #label
df['SellerG'].unique()
df['CouncilArea'].unique()
df['Regionname'].unique() #label
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
label = LabelEncoder()
df['Type_encoded'] = label.fit_transform(df['Type'])
```

```python
df
```

```python
df['Method_encoded'] = label.fit_transform(df['Method'])
```

```python
df
```

```python
df['Regionname_encoded'] = label.fit_transform(df['Regionname'])
```

```python
df.drop(columns = ['Type' , 'Method' , 'Regionname'] , axis = 1 , inplace = True)
```

#This is the shortcut

```python
#onehot = pd.get_dummies(df, columns = ['Suburb', 'SellerG', 'CouncilArea'], dtype = int)
# onehot
```

```python
onehot = OneHotEncoder(sparse_output = False)
suburb_data = onehot.fit_transform(df[['Suburb']])
suburb_data
```

```python
s_name = onehot.get_feature_names_out()
df1 = pd.DataFrame(suburb_data,columns = s_name)
df1
```

```python
councilarea_data = onehot.fit_transform(df[['CouncilArea']])
councilarea_data
```

```python
ca_name = onehot.get_feature_names_out()
df2 = pd.DataFrame(councilarea_data,columns = ca_name)
df2
```

```python
seller_data = onehot.fit_transform(df[['SellerG']])
seller_data
```

```python
sl_name = onehot.get_feature_names_out()
df3 = pd.DataFrame(seller_data,columns = sl_name)
df3
```

```python
df = pd.concat([df, df1 ,df2 , df3], axis=1)
df
```

```python
df.drop(columns = ['Suburb' , 'CouncilArea' , 'SellerG'] , axis = 1 , inplace = True)
```

```python
df
```

```python
df.shape
```

```python
x = df.drop('Price' , axis = 1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.7 , random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
model.score(x_train , y_train)
```

```python
model.score(x_test , y_test)
```

```python
from sklearn.metrics import r2_score , mean_squared_error , mean_absolute_error
```

```python
r2 = r2_score(y_test , y_pred)
r2
```
