```python
import pandas as pd
from sklearn.datasets import load_iris, fetch_california_housing
from sklearn.tree import DecisionTreeClassifier , DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score , mean_squared_error , mean_absolute_error
```

# Decision Tree Classifier

```python
data = load_iris()
df = pd.DataFrame(data.data , columns = data.feature_names)
df['species'] = data.target
df
```

```python
x = df.drop('species' , axis = 1)
y = df.species
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model = DecisionTreeClassifier(max_depth = 5)
model.fit(x_train , y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
print('Training score: ', model.score(x_train,y_train))
```

```python
print('Testing score: ', model.score(x_test,y_test))
```

# Decision Tree Regressor

```python
rdata = fetch_california_housing()
df = pd.DataFrame(rdata.data , columns = rdata.feature_names)
df['target'] = rdata.target
df
```

```python
x = df.drop('target' , axis = 1)
y = df.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
rmodel = DecisionTreeRegressor(max_depth = 5)
rmodel.fit(x_train , y_train)
```

```python
y_pred = rmodel.predict(x_test)
y_pred
```

```python
print('Training score: ', rmodel.score(x_train,y_train))
```

```python
print('Testing score: ', rmodel.score(x_test,y_test))
```

```python
print("r2 score:" , r2_score(y_test , y_pred))
```

```python
print("mae score:" , mean_absolute_error(y_test , y_pred))
```

```python
print("mse score:" , mean_squared_error(y_test , y_pred))
```
