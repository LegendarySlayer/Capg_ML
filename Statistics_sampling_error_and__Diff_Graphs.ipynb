```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/winequalityN.csv")
df
```

```python
pop = df['alcohol']
sample_size = 500
sample = df['alcohol'].sample(sample_size)
sample_mean = sample.mean()
pop_mean = pop.mean()
error = pop_mean - sample_mean
print(error)
print(sample_mean)
print(pop_mean)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(range(1,sample_size + 1),sample)
plt.axhline(pop.mean() , color = 'red' , label = 'population mean')
plt.axhline(sample.mean() , color = 'blue' , label = 'sample mean')
plt.legend()
plt.show()
```

```python
pop_white = df.groupby('type').get_group('white')['alcohol']
# pop_red = df.groupby('type').get_group('red')['alcohol']
sample_size = 500
sample_w = pop_white.sample(sample_size)
sample_mean_w = sample_w.mean()
pop_mean_w = pop_white.mean()
error1 = pop_mean_w - sample_mean_w
print(error1)
# print(sample_mean)
# print(pop_mean)
```

```python
plt.scatter(range(1,sample_size + 1),sample_w)
plt.axhline(pop_white.mean() , color = 'red' , label = 'population mean of white')
plt.axhline(sample_w.mean() , color = 'blue' , label = 'sample mean')
plt.legend()
plt.show()
```

```python
pop_red = df.groupby('type').get_group('red')['alcohol']
sample_size = 500
sample_r = pop_red.sample(sample_size)
sample_mean_r = sample_r.mean()
pop_mean_r = pop_red.mean()
error2 = pop_mean_r - sample_mean_r
print(error2)
```

```python
plt.scatter(range(1,sample_size + 1),sample_r)
plt.axhline(pop_red.mean() , color = 'red' , label = 'population mean of red ')
plt.axhline(sample_r.mean() , color = 'blue' , label = 'sample mean')
plt.legend()
plt.show()
```

```python
sample_age = df_new['Age'].sample(1000)
plt.figure(figsize=(15,5))
plt.plot(sample_age.values, linewidth=1)
plt.title("Line Plot of Age (Sample 1000 rows)")
plt.xlabel("Sample Index")
plt.ylabel("Age")
plt.show()
```

```python
#Line Chart
sample_age = df_new['Age'].sample(1000)
plt.figure(figsize=(15,5))
plt.plot(sample_age.values, linewidth=1)
plt.axhline(sample_age.mean(), color='red', linestyle='--', label='Average Age')
plt.title("Line Plot of Age with Average Line")
plt.xlabel("Sample Index")
plt.ylabel("Age")
plt.legend()
plt.show()
```

```python
#Histogram chart
plt.figure(figsize=(8,5))
plt.hist(df_new['Age'], bins=25, color='skyblue', edgecolor='black')
plt.title("Distribution of Athlete Ages")
plt.xlabel("Age")
plt.ylabel("Number of Athletes")
plt.show()
```

```python
#Bar Graph
plt.figure(figsize=(7,5))
df_new['Sex'].value_counts().plot(kind='bar', color=['blue','pink'])
plt.title("Number of Male and Female Athletes")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#Scatter Map
plt.figure(figsize=(7,5))
plt.scatter(df_new['Height'], df_new['Weight'], alpha=0.5)
plt.title("Relationship between Height and Weight")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
#Heat Map
plt.figure(figsize=(6,5))
sns.heatmap(df_new[['Age','Height','Weight']].corr(),
            annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#pie chart
counts = df_new['Team'].value_counts()
top5 = counts.head(5)
others = counts[5:].sum()
top5['Others'] = others
plt.figure(figsize=(6,6))
plt.pie(top5, labels = top5.index, autopct='%1.1f%%')
plt.title("Top 5 Teams Distribution")
plt.show()
```
